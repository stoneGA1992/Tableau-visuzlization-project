```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python

april2018 = pd.read_csv('April_2018.csv')
april2019 = pd.read_csv('April_2019.csv')
april2020 = pd.read_csv('April_2020.csv')
```

```python
april2018 
```

```python
#Add the Year and month columns to 2018
april2018 = april2018.assign(year = '2018')
april2018 = april2018.assign(month = 'April')
april2018
```

```python
#Add the Year and month columns to 2019
april2019 = april2019.assign(year = '2019')
april2019 = april2019.assign(month = 'April')
april2019
```

```python
#Add the Year and month columns to 2020
april2020 = april2020.assign(year = '2020')
april2020 = april2020.assign(month = 'April')
april2020
```

```python
#Combine all into one spreadsheet

combined_data = april2018.append([april2019, april2020], ignore_index = True)

combined_data
```
